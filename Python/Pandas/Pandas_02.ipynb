```python
import matplotlib.pyplot as plt#Visualisation
import seaborn as sns #its used for Statistical visualization

#Keep the graph in proper shape
%matplotlib inline
plt.rcParams['figure.figsize'] = 6,2 #6: height of graph, 2: width of graph

#to ignore warning messages
import warnings
warnings.filterwarnings('ignore')
```

```python
vis1 = sns.distplot(df['InternetUsers'])#Distributions Plot
#Here since we are using only 1 variable i.e InternetUsers its called Univariate analysis
```

```python
vis2 = sns.displot(df['InternetUsers'])#Distributions Plot
```

```python
vis3 = sns.distplot(df['InternetUsers'],bins = 10)#bins=10 means we willget 10 bars
```

```python
#Bivariate Graph : Plot the graph using 2 variable called Bi-Variate Graph
#Multi-variate Graph : Plot the graph using more than 2 variable called multi-Variate Graph
vis4 = sns.boxplot(data = df, x = "IncomeGroup", y = "BirthRate")
```

```python
#Outlier is datapoint which is far from other observations of the data point.
#in 1st graph min line(below horizontal line) is at 7.9, max i around 21 but we have 1 circle at 35 which is outlier
#Outlier AKA anomaly detection
#Blue box is interquartile
#Within box we have line is called median
#lower line of box is lower quartile
#Upper line of boz is upper quartile 
#Standing line between (min to lower quartile & upper quartile to max)is called whisker 
```

```python
#LMPLot Liner Model y = mx+c
vis5 = sns.lmplot(data = df, x = "InternetUsers", y = "BirthRate")
```

```python
vis6 = sns.lmplot(data = df, x = "InternetUsers", y = "BirthRate", fit_reg = False)
```

```python
#To add color to a attribute/column
vis6 = sns.lmplot(data = df, x = "InternetUsers", y = "BirthRate", fit_reg = False, hue = "IncomeGroup")
plt.show()
```

```python
#To add color to a attribute/column
vis6 = sns.lmplot(data = df, x = "InternetUsers", y = "BirthRate", fit_reg = True, hue = "IncomeGroup")
plt.show()
```
